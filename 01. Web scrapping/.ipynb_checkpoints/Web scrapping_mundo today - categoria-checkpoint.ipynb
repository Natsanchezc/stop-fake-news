```python
# Creamos una función que obtenga los links a las noticias a partir de un enlace de Elmundotoday correspondiente a las noticias de un mes
def obtener_links_emt(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")
    elementos = soup.find_all("a", {"rel":"bookmark"})
    links =[]
    for item in elementos:
        links.append(item["href"])
    return(links)
```

```python
url = "https://www.elmundotoday.com/2023/07/"
obtener_links_emt(url)
```

```python
# Obtener lista de links a elmundotoday por meses para posteriormente iterar sobre ellos y obtener la lista de links de cada noticia

url = "https://www.elmundotoday.com/2010/00/"
lista_url_años=[] # lista de links de cada año
for _ in range(13):
    patron = r"(https://www\.elmundotoday\.com/)(\d{4})(/)(\d+/)"
    url = re.sub(patron, lambda m: m.group(1) + str(int(m.group(2)) + 1).zfill(4) + m.group(3) + m.group(4), url)
    lista_url_años.append(url)

lista_url_meses=[] 
for url in lista_url_años: # itero sobre la lista de links de cada año para obtener los meses
    for _ in range(12):
        patron = r"(https://www\.elmundotoday\.com/)(\d{4}/)(\d+)(/)"
        url = re.sub(patron, lambda m: m.group(1) + m.group(2) + str(int(m.group(3)) + 1).zfill(2) + m.group(4), url)
        lista_url_meses.append(url)
lista_url_meses    
```

```python
# Eliminamos los últimos 5 meses de este año
lista_url_meses=lista_url_meses[:-5]
```

```python
len(lista_url_meses)
```

```python
links_noticias_emt=[]
for url in lista_url_meses:
    links_noticias_emt.append(obtener_links_emt(url))
links_noticias_emt
```

Definir una función que obtenga el título, subtítulo, cuerpo y fecha de cada link

```python
# le paso la variable links que será una lista de hipervínculos de noticias de Elmundotoday
# recorro la lista de links para obtener la noticia de cada uno

def obtener_noticias_emt(links):
    
    lista_fechas=[]
    lista_titulos=[]
    lista_subtitulos=[]
    lista_textos=[]
    
    for url in links:
        response = requests.get(url) # peticion http
        soup = BeautifulSoup(response.text, "html.parser")
        
        #Título
        titulo = soup.find("h1").text 
        lista_titulos.append(titulo)
        
        #Subtítulo
        soup_subtitulo = soup.find_all("div", {"class":"tdb-block-inner td-fix-index"})[7].find("p")
        if soup_subtitulo == None:
            lista_subtitulos.append("NaN")
        else:
            subtitulo = soup.find_all("div", {"class":"tdb-block-inner td-fix-index"})[7].find("p").text 
            lista_subtitulos.append(subtitulo)
            
        #Fecha
        str(soup.find("meta", {"property":"article:published_time"}))
        soup_fecha = str(soup.find("meta", {"property":"article:published_time"}))
        resultado = re.findall(r'"([^"]*)"', soup_fecha)
        time_publicacion_noticia = ''.join(resultado)
        fecha = time_publicacion_noticia[:10]
        lista_fechas.append(fecha)
        
        #Cuerpo de la noticia
        try:
            cuerpo=soup.find_all("div", {"class":"tdb-block-inner td-fix-index"})[11].text
            lista_textos.append(cuerpo)

        except Exception:
            lista_textos.append("NaN")
            
        
    return(lista_fechas, lista_titulos, lista_subtitulos, lista_textos)
```

```python
# Función para crear csv de noticias de Elmundotoday

def crear_df_noticias_emt(lista_urls, pos):
    
    list(obtener_noticias_emt(lista_urls))

    fechas=obtener_noticias_emt(lista_urls)[0]
    titulos=obtener_noticias_emt(lista_urls)[1]
    subtitulos=obtener_noticias_emt(lista_urls)[2]
    cuerpos=obtener_noticias_emt(lista_urls)[3]
    
    df_emt = pd.DataFrame({"Periódico":"El Mundo Today", 
                                   "Hipervínculo":lista_urls, 
                                   "Fecha publicación":fechas, 
                                   "Titular":titulos, 
                                   "Subtítulo":subtitulos,
                                   "Target":1,
                                   "Cuerpo":cuerpos})
    df_emt.to_csv("data_emt_"+ str(pos) + ".csv", index=False)
    
```
