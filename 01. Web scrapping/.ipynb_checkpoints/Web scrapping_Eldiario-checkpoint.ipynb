```python
lista_categorias = ["politica", "internacional", "deportes", "sociedad", "economia", "cultura", "educacion", "clima", "desalambre"]
```

```python
lista_links_categorias=[]
for categoria in lista_categorias:
    lista_links_categorias.append(r"https://www.eldiario.es/" + categoria + "/")
```

```python
lista_links_categorias
```

```python
def obtener_links_eldiarioes(lista_urls):
    
    lista_links=[]
    lista_links_final=[]
    
    for url in lista_urls:
        response = requests.get(url)
        soup = BeautifulSoup(response.text, "html.parser")
        
        for item in soup.find_all("h2"):
            lista_links.append(item.find("a")["href"])
            
    for link in lista_links:
        lista_links_final.append(link)
        
    return lista_links_final 
```

```python
# Filtramos la lista de links que hemos obtenido para que sean correctos 

def eliminar_links_nosoneldiario(lista, caracteres):
    return list(filter(lambda x: any(c in x for c in caracteres), lista))

caracteres = ['https://www.eldiario.es/']

lista_links_filtrada = eliminar_links_nosoneldiario(lista_links_ed, caracteres) # LISTA FINAL DE LINKS
len(lista_links_filtrada) 
```

```python
# Función obtener título, subtitulo, cuerpo, categoría y fecha de cada noticia
```

```python
def obtener_noticias_eldiarioes(links):
    
    lista_fechas=[]
    lista_titulos=[]
    lista_subtitulos=[]
    lista_textos=[]
    lista_categorias=[]
    
    for url in links:
        response = requests.get(url) # peticion http
        soup = BeautifulSoup(response.text, "html.parser")
        
        #Fecha
        soup_fecha=str(soup.find("time", {"class":"date"}))
        patron = r'202\d{1}-\d{2}-\d{2}'
        try:
            fecha = re.search(patron, soup_fecha).group()
            print(fecha)
            lista_fechas.append(fecha)
        except Exception:
            print("NaN")
            lista_fechas.append("NaN")
            
        #Título
        try:
            titulo = (soup.find("h1").text).strip()
            print(titulo)
            lista_titulos.append(titulo)
        except Exception:
            lista_titulos.append("NaN")
            
        #Subtítulo
        try:    
            subtitulo = (soup.find("h2").text).strip()
            lista_subtitulos.append(subtitulo)
        except Exception:
            lista_subtitulos.append("NaN")
             
        #Cuerpo de la noticia
        cuerpo = []
        try:
            for item in soup.find_all("p", {"class":"article-text"}):
                parrafo = (item.text).strip()
                cuerpo.append(parrafo)
            cuerpo_combinado = ''.join(str(elemento) for elemento in cuerpo)
            cuerpo_filtrado = cuerpo_combinado.replace("\xa0", " ")
            lista_textos.append(cuerpo_filtrado)
        except Exception:
            lista_textos.append("NaN")
            
        #Categoria
        patron = r"https://www.eldiario.es/(.*?)/" 
        categoria_search = re.findall(patron, url)
        categoria = ''.join(categoria_search)
        lista_categorias.append(categoria)
        
    return [lista_fechas, lista_titulos, lista_subtitulos, lista_textos, lista_categorias]
```

```python
# Función para crear csv de noticias de ElDiario.es

def crear_csv_noticias_eldiarioes(lista_urls, pos):

    fechas=obtener_noticias_eldiarioes(lista_urls)[0]
    titulos=obtener_noticias_eldiarioes(lista_urls)[1]
    subtitulos=obtener_noticias_eldiarioes(lista_urls)[2]
    cuerpos=obtener_noticias_eldiarioes(lista_urls)[3]
    categorias=obtener_noticias_eldiarioes(lista_urls)[4]
    
    df_ed = pd.DataFrame({"Periódico":"ElDiario.es", 
                                   "Hipervínculo":lista_urls, 
                                   "Fecha publicación":fechas, 
                                   "Titular":titulos, 
                                   "Subtítulo":subtitulos,
                                   "Cuerpo":cuerpos,
                         "Categoría":categorias,
                         "Target":0})
    
    df_ed.to_csv("data_eldiarioes_"+ str(pos) + ".csv", index=False)
```
