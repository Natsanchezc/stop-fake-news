```python
# Creamos una función que obtenga los links a las noticias a partir de un enlace de Elmundotoday correspondiente a las noticias de un mes
def obtener_links_elmundo(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")
    elementos = soup.find_all("a", {"class":"ue-c-cover-content__link"})
    links =[]
    for item in elementos:
        links.append(item["href"])
    return(links)
```

```python
url = 'https://www.elmundo.es/elmundo/hemeroteca/2020/06/21/m/index.html'
obtener_links_elmundo(url)
```

```python
lista_links=obtener_links_elmundo(url)
len(lista_links)
```

```python
# Filtramos la lista de links que hemos obtenido para quedarnos con los más objetivos

def eliminar_links_nosonelmundo(lista):
    caracteres = ['https://www.elmundo.es/']
    return list(filter(lambda x: any(c in x for c in caracteres), lista))
```

```python
def eliminar_links_opinion(lista):
    caracteres = ['https://www.elmundo.es/opinion/']
    return list(filter(lambda x: not any(c in x for c in caracteres), lista))
```

```python
lista_links_filtrada_2 = eliminar_links_opinion(lista_links_filtrada) # LISTA FINAL DE LINKS
len(lista_links_filtrada_2) 
```

```python
# Obtener lista de links a elmundo por fecha para posteriormente iterar sobre ellos y obtener la lista de links de cada noticia

url = 'https://www.elmundo.es/elmundo/hemeroteca/2022/00/00/m/index.html'
lista_url_meses=[] # lista de links de cada año
for _ in range(12):
    patron = r"(https://www.elmundo.es/elmundo/hemeroteca/2022/)(\d{2})(/00/m/index.html)"
    url = re.sub(patron, lambda m: m.group(1) + str(int(m.group(2)) + 1).zfill(2) + m.group(3), url)
    lista_url_meses.append(url)

lista_urls_año2022=[] 
for url in lista_url_meses: # itero sobre la lista de links de cada año para obtener los meses
    for _ in range(25):
        patron = r"(https://www.elmundo.es/elmundo/hemeroteca/2022/)(\d{2}/)(\d{2})(/m/index.html)"
        url = re.sub(patron, lambda m: m.group(1) + m.group(2) + str(int(m.group(3)) + 1).zfill(2) + m.group(4), url)
        lista_urls_año2022.append(url)
lista_urls_año2022  
```

```python
len(lista_urls_año2022)
```

```python
# Obtenemos los urls a las noticias de cada link de cada día del 2022
lista_links=[]
for link in lista_urls_año2022:
    lista_links.append(obtener_links_elmundo(link))
urls_noticias_elmundo = [elemento for sublista in lista_links for elemento in sublista]  
urls_noticias_elmundo
```

```python
len(urls_noticias_elmundo)
```

```python
# Filtramos para quedarnos con las noticias más objetivas
urls_noticias_filtrada = eliminar_links_nosonelmundo(urls_noticias_elmundo)
len(urls_noticias_filtrada)
```

```python
# Filtramos las noticias de opinión
urls_noticias_filtrada2 = eliminar_links_opinion(urls_noticias_filtrada)
len(urls_noticias_filtrada2)
```

```python
# Filtramos noticias que parecen subjetivas (categorías: crónica y españa)
def eliminar_links_categoria(lista):
    caracteres = ['https://www.elmundo.es/cronica/', 'https://www.elmundo.es/espana/', 
                  'https://www.elmundo.es/album/', 'https://www.elmundo.es/podcasts/',
                 'https://www.elmundo.es/eventos/en-directo/']
    return list(filter(lambda x: not any(c in x for c in caracteres), lista))
```

```python
urls_noticias_filtrada3 = eliminar_links_categoria(urls_noticias_filtrada2)
len(urls_noticias_filtrada3)
```

```python
# Función para obtener los datos de todas las noticias
```

```python
def obtener_noticias_elmundo(links):
    
    lista_fechas=[]
    lista_titulos=[]
    lista_subtitulos=[]
    lista_textos=[]
    lista_categorias=[]
    
    for url in links:
        response = requests.get(url) # peticion http
        soup = BeautifulSoup(response.text, "html.parser")
        
        #Fecha
        soup_fecha=str(soup.find("time"))
        patron = r'202\d{1}-\d{2}-\d{2}'
        try:
            fecha = re.search(patron, soup_fecha).group()
            print(fecha)
            lista_fechas.append(fecha)
        except Exception:
            print("NaN")
            lista_fechas.append("NaN")
            
        #Título
        titulo = soup.find("h1").text
        print(titulo)
        lista_titulos.append(titulo)
            
        #Subtítulo
        try:    
            subtitulo = soup.find("p", {'class':"ue-c-article__standfirst"}).text
            lista_subtitulos.append(subtitulo)
        except Exception:
            lista_subtitulos.append("NaN")
             
        #Cuerpo de la noticia
        try:    
            cuerpo = []
            parrafos = soup.find_all("div", {"class":"ue-l-article__body ue-c-article__body"})[0].find_all("p")
            for item in parrafos:
                cuerpo.append(item.text)
            cuerpo_combinado = ''.join(str(elemento) for elemento in cuerpo[:-1])
            lista_textos.append(cuerpo_combinado)
        except Exception:
            lista_textos.append("NaN")
            
        #Categoria
        patron = r"https://www.elmundo.es/(.*?)/" 
        categoria_search = re.findall(patron, url)
        categoria = ''.join(categoria_search)
        lista_categorias.append(categoria)
        
    return [lista_fechas, lista_titulos, lista_subtitulos, lista_textos, lista_categorias]
```

```python
noticias = obtener_noticias_elmundo(urls_noticias_filtrada3[:200])
```

```python
# Función para crear csv de noticias de ElMundo

def crear_csv_noticias_elmundo(lista_urls, pos):

    fechas=obtener_noticias_elmundo(lista_urls)[0]
    titulos=obtener_noticias_elmundo(lista_urls)[1]
    subtitulos=obtener_noticias_elmundo(lista_urls)[2]
    cuerpos=obtener_noticias_elmundo(lista_urls)[3]
    categorias=obtener_noticias_elmundo(lista_urls)[4]
    
    df = pd.DataFrame({"Periódico":"ElMundo", 
                                   "Hipervínculo":lista_urls, 
                                   "Fecha publicación":fechas, 
                                   "Titular":titulos, 
                                   "Subtítulo":subtitulos,
                                   "Cuerpo":cuerpos,
                         "Categoría":categorias,
                         "Target":0})
    
    df.to_csv("data_elmundo_"+ str(pos) + ".csv", index=False)
```
