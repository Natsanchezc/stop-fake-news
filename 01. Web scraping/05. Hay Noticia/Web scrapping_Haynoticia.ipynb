```python
url_paginas=[] # lista de links de cada año
for url in links_categorias:
    for _ in range(10):
        patron = r"(https://haynoticia.es/[a-z]{3,}/page/)(\d{1})(/)"
        url = re.sub(patron, lambda m: m.group(1) + str(int(m.group(2)) + 1) + m.group(3), url)
        url_paginas.append(url)
url_paginas
```

```python
def obtener_links_haynoticia(lista_urls):
    
    lista_links=[]
    lista_categoria=[]
    
    for url in lista_urls:
        response = requests.get(url)
        soup = BeautifulSoup(response.text, "html.parser")
        try:
            for item in soup.find_all("h2", {"class":"entry-title"}):
                lista_links.append(item.find("a")["href"])
                patron = r"https://haynoticia.es/(.*?)/" 
                categoria_search = re.findall(patron, url)
                categoria = ''.join(categoria_search)
                lista_categoria.append(categoria)
        except Exception:
            lista_links.append("NaN")
        
    
    return [lista_links, lista_categoria]
```

```python
def obtener_noticias_haynoticia(links):
    
    lista_fechas=[]
    lista_titulos=[]
    lista_textos=[]

    
    for url in links:
        response = requests.get(url) # peticion http
        soup = BeautifulSoup(response.text, "html.parser")
        
        #Fecha
        soup_fecha=soup.find("time", {"class":"entry-date published"})
        patron = r'20\d{2}-\d{2}-\d{2}'
        try:
            fecha = re.search(patron, str(soup_fecha)).group()
            print(fecha)
            lista_fechas.append(fecha)
        except Exception:
            print("NaN")
            lista_fechas.append(np.nan)
            
        #Título
        titulo = soup.find("h1", {"class":"entry-title"}).text.strip()
        print(titulo)
        lista_titulos.append(titulo)
             
        #Cuerpo de la noticia
        cuerpo = soup.find("div", {"class":"entry-content clearfix"}).find_all("p")
        cuerpo_filtrado = []
        for item in cuerpo:
            cuerpo_filtrado.append(item.text)
        cuerpo_combinado = ''.join(str(elemento) for elemento in cuerpo_filtrado)
        lista_textos.append(cuerpo_combinado)
          
        
    return [lista_fechas, lista_titulos, lista_textos]
```

```python
# Función para crear csv de noticias de ElDiario.es

def crear_csv_noticias_haynoticia(lista_urls, pos, lista_categorias):

    fechas=obtener_noticias_haynoticia(lista_urls)[0]
    titulos=obtener_noticias_haynoticia(lista_urls)[1]
    cuerpos=obtener_noticias_haynoticia(lista_urls)[2]
    
    df_hn = pd.DataFrame({"Periódico":"HayNoticia.es", 
                                   "Hipervínculo":lista_urls, 
                                   "Fecha publicación":fechas, 
                                   "Titular":titulos, 
                                   "Subtítulo":np.nan,
                                   "Cuerpo":cuerpos,
                         "Categoría":lista_categorias,
                         "Target":1})
    
    df_hn.to_csv("data_haynoticia_"+ str(pos) + ".csv", index=False)
```
