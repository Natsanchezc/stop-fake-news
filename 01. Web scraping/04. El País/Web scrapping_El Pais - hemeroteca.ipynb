```python
def eliminar_elementos(lista):
    caracteres=["opinion","publi",'https://', '/juegos/', '/especiales/', '/ideas/', '/videos/', '/babelia/']
    return list(filter(lambda x: not any(c in x for c in caracteres), lista))
```

```python
def obtener_urls_noticias_elpais(urls):
    lista_links=[]   
    for url in urls:
        response = requests.get(url)
        soup = BeautifulSoup(response.text, "html.parser")
        for item in soup.find_all("h2"):
            lista_links.append(item.find("a")["href"])
    # Filtramos la lista 
    lista_filtrada = eliminar_elementos(lista_links)
    
    # Añadir 'https://elpais.com' al comienzo de cada hipervínculo
    caracteres = 'https://elpais.com'
    lista_modificada = [caracteres + cadena for cadena in lista_filtrada]
    # Eliminar los links repetidos
    lista_sin_repetidos = list(set(lista_modificada))
    return lista_sin_repetidos
```

```python
# Obtener lista de links a ElPais por fecha para posteriormente iterar sobre ellos y obtener la lista de links de cada noticia

url = 'https://static.elpais.com/hemeroteca/elpais/2022/00/00/m/portada.html'
lista_url_meses=[] # lista de links de cada año
for _ in range(12):
    patron = r"(https://static.elpais.com/hemeroteca/elpais/2022/)(\d{2})(/00/m/portada.html)"
    url = re.sub(patron, lambda m: m.group(1) + str(int(m.group(2)) + 1).zfill(2) + m.group(3), url)
    lista_url_meses.append(url)

lista_urls_año2022=[] 
for url in lista_url_meses: # itero sobre la lista de links de cada año para obtener los meses
    for _ in range(25):
        patron = r"(https://static.elpais.com/hemeroteca/elpais/2022/)(\d{2}/)(\d{2})(/m/portada.html)"
        url = re.sub(patron, lambda m: m.group(1) + m.group(2) + str(int(m.group(3)) + 1).zfill(2) + m.group(4), url)
        lista_urls_año2022.append(url)
lista_urls_año2022 
```

```python
# Obtenemos los urls a las noticias de cada link de cada día del 2022
lista_urls_noticias = obtener_urls_noticias_elpais(lista_urls_año2022)
```

```python
#Función para obtener todos los datos de cada noticia

def obtener_noticias_elpais(links):
    
    lista_fechas=[]
    lista_titulos=[]
    lista_subtitulos=[]
    lista_textos=[]
    lista_categorias=[]
    
    for url in links:
        response = requests.get(url) # peticion http
        soup = BeautifulSoup(response.text, "html.parser")
        
        #Fecha
        try:
            patron = r'20\d{2}-\d{2}-\d{2}'
            fecha = re.search(patron, url).group()
            print(fecha)
            lista_fechas.append(fecha)
        except Exception:
            lista_fechas.append(np.nan)
        
        #Título
        try:
            titulo = soup.find("h1", {"class":"a_t"}).text 
            lista_titulos.append(titulo)
            print(titulo)
        except Exception:
            lista_titulos.append(np.nan)
        
        #Subtítulo
        try:
            subtitulo = soup.find("h2", {"class":"a_st"}).text
            lista_subtitulos.append(subtitulo)
        except Exception:
            lista_subtitulos.append(np.nan)
             
        #Cuerpo de la noticia
        try: 
            cuerpo=[]
            soup_cuerpo=soup.find("div", {"class":"a_c clearfix"}).find_all('p')
            for parrafo in soup_cuerpo:
                cuerpo.append(parrafo.text)
            cuerpo_combinado = ''.join(str(elemento) for elemento in cuerpo)
            lista_textos.append(cuerpo_combinado)
            
        except Exception:
            lista_textos.append(np.nan)
            
        #Categoría
        patron = r"https://elpais.com/(.*?)/" 
        categoria_search = re.findall(patron, url)
        categoria = ''.join(categoria_search)
        lista_categorias.append(categoria)
        
    return [lista_fechas, lista_titulos, lista_subtitulos, lista_textos, lista_categorias]
```

```python
# Función para crear csv de noticias de ElPais

def crear_csv_noticias_elpais(lista_urls, pos):

    fechas=obtener_noticias_elpais(lista_urls)[0]
    titulos=obtener_noticias_elpais(lista_urls)[1]
    subtitulos=obtener_noticias_elpais(lista_urls)[2]
    cuerpos=obtener_noticias_elpais(lista_urls)[3]
    categorias=obtener_noticias_elpais(lista_urls)[4] 
    
    df_ep = pd.DataFrame({"Periódico":"El País", 
                                   "Hipervínculo":lista_urls, 
                                   "Fecha publicación":fechas, 
                                   "Titular":titulos, 
                                   "Subtítulo":subtitulos,
                          "Cuerpo":cuerpos,
                          "Categoría":categorias,
                                   "Target":0,})
    df_ep.to_csv("data_elpais_" + str(pos) + ".csv", index=False)
```
